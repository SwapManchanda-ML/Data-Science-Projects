# Metal Consumption Rates

```python
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
import matplotlib.style as style 
sns.set_style("darkgrid")
sns.set_context("poster")

df = pd.read_excel(r'Raw_Data.xlsx',sheet_name='Figure1_CleanData')
df.head(15)
```

```python
df_ordered = df.sort_values(by = 'HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT TODAYS RATE (YRS)')
```

```python

plt.figure(figsize=(17,17))
plt.hlines(y = df_ordered['Material'], xmin = df_ordered['HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT THE AVERAGE US RATE (YRS)'],
           xmax = df_ordered['HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT TODAYS RATE (YRS)'],
           color='black', alpha=0.5)
plt.scatter(df_ordered['HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT THE AVERAGE US RATE (YRS)'],  
            df_ordered['Material'], color='red', alpha=0.5, label='Years left if everyone in the world consumed at the same rate an average person in the USA consumed in 2006')
plt.scatter(df_ordered['HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT TODAYS RATE (YRS)'],
            df_ordered['Material'],marker='s', color='green', alpha=0.5 , label='Years left if everyone in the world consumed at the rate they did in 2006')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.075), shadow=True, ncol=1)
plt.xlabel('Years')
plt.ylabel('Material')

for x,y in zip(df_ordered['HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT THE AVERAGE US RATE (YRS)'],df_ordered['Material']):

    label = "{:.2f}".format(x)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='right',
                 color = 'red') # horizontal alignment can be left, right or center
for x,y in zip(df_ordered['HOW MANY YEARS LEFT IF THE WORLD CONSUMES AT TODAYS RATE (YRS)'],df_ordered['Material']):

    label = "{:.2f}".format(x)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='left',
                 color = 'green') # horizontal alignment can be left, right or center
plt.rcParams["figure.facecolor"] = "peachpuff"
plt.title('Years Left After 2006 Until Depletion',fontsize=25)   
plt.show()


```

```python
df_pie = pd.read_excel(r'Raw_Data.xlsx',sheet_name='Sheet4')
df_pie.head(15)
```

```python
fig, axes = plt.subplots(5, 3, figsize=(18, 18))

txt="Values rounded to the nearest Tonne."

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format

for i, (idx, row) in enumerate(df_pie.set_index('Material').iterrows()):
    ax = axes[i // 3, i % 3]
    row = row[row.gt(row.sum() * 0.01)]
    test = ax.pie(row, startangle=115,autopct = autopct_format(row/1000),pctdistance=1.75,radius=1.2, colors =['black','lightseagreen'])
    plt.setp(test[0], width=0.2, edgecolor='white')
    
    kwargs = dict(size=17, fontweight='bold', va='center')
    ax.text(0, 0, idx, ha='center', **kwargs)
    
fig.subplots_adjust(wspace=.3,hspace=0.5)
fig.text(0, 0, txt, ha='left')
fig.legend(test[0], df_pie.columns[1:3], loc='lower right', bbox_to_anchor=(0., 0, 1., .1), shadow=True, ncol=1)
fig.suptitle('Recycled and Mined Raw Material Production In 2006', size=25)
 
```
